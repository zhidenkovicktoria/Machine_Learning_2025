### 3. Построение Байесовсой сети

#### Автоматическое обучение структуры (HillClimbSearch + BIC)

````python
import pandas as pd
import numpy as np
from pgmpy.estimators import HillClimbSearch, BicScore, BayesianEstimator, MaximumLikelihoodEstimator
from pgmpy.models import BayesianNetwork
from pgmpy.inference import VariableElimination
import matplotlib.pyplot as plt
import networkx as nx
````

````python
# HillClimbSearch — это жадный алгоритм, который ищет наилучшую структуру
# BicScore (Bayesian Information Criterion) — штрафует за сложность модели
hc = HillClimbSearch(data)
best_model_structure = hc.estimate(scoring_method=BicScore(data))

model = BayesianNetwork(best_model_structure.edges())

print(f"Найдено узлов: {len(model.nodes())}")
print(f"Найдено рёбер: {len(model.edges())}")

# Выводим найденные рёбра
print("\nРёбра байесовской сети:")
for edge in best_model_structure.edges():
    print(edge)
````

### 3. Получение вероятностей

Мы хотим получить вероятностные оценки переходов из одного состояние в другое. Для этого необходимо Байесовскую сеть подогнать по данные и получить таблички (CPT).

Conditional Probability Table (CPT) [Таблица Условных Вероятностей (CPT)] — это ключевой элемент, определяющий параметры Байесовской сети.

* Что это: Это таблица, которая содержит вероятность каждого возможного состояния узла (переменной) при условии всех возможных комбинаций состояний его родительских узлов (тех, от которых к нему идут стрелки/ребра).
* Роль: CPTs количественно определяют зависимости и силу связей между переменными, позволяя сети выполнять логический вывод и прогнозирование.
* Простой пример: В CPT для узла "Кашель", родительским узлом которого является "Грипп", будет указана вероятность кашля, если грипп есть (например, 90%), и вероятность кашля, если гриппа нет (например, 5%).

Чтобы получить значения вероятности переходов из разных узлов, нужно оценить эти вероятности. `HillClimbSearch` ищет только структуру, а `MaximumLikelihoodEstimator` и `BayesianEstimator` позволяют оценить вероятность.

````python
from pgmpy.estimators import MaximumLikelihoodEstimator, BayesianEstimator
from pgmpy.inference import VariableElimination

model.fit(data, estimator = BayesianEstimator, prior_type = 'BDeu', equivalent_sample_size = 10)
````

````python
# Выводим ВСЕ таблицы CPT по одной (с названием узла)
for cpd in model.cpds:
    print(f"\n{'='*60}")
    print(f"Таблица CPT для узла: {cpd.variable}  (всего состояний: {cpd.variable_card})")
    if len(cpd.variables) > 1:
        print(f"Зависит от: {', '.join(cpd.variables[:-1])}")
    else:
        print("Этот узел не имеет родителей (корневой)")
    print(f"{'='*60}")
    print(cpd)
    print("\n")
````

Ключевая таблица: как запах определяет ядовитость

````python
# Самая важная таблица в датасете грибов
cpd_class = model.get_cpds('class')
print("P(class | odor) — определяет съедобность/ядовитость:")
print(cpd_class)
````

Узел class зависит от odor и stalk-shape. При определённых комбинациях значений этих признаков вероятность принадлежности к классу съедобных или ядовитых грибов составляет 100%. Это означает, что построенная байесовская сеть способна безошибочно определять токсичность гриба по его морфологическим признакам.

### 4. Визуализация сети через библиотеку `networkx`

### 5. Посмотрим пример инференса

````python
# Точный вероятностный вывод (Inference) с помощью VariableElimination 

from pgmpy.inference import VariableElimination

infer = VariableElimination(model)   

print("═" * 80)
print("ПРИМЕРЫ ВЕРОЯТНОСТНОГО ВЫВОДА: определение съедобности гриба")
print("═" * 80)

def predict_mushroom(evidence):
    result = infer.query(variables=['class'], evidence=evidence)
    prob_edible = result.values[0]
    prob_poisonous = result.values[1]
    
    print(f"\nУсловия: {evidence}")
    print(f"→ P(съедобный)  = {prob_edible:8.4%}")
    print(f"→ P(ядовитый)   = {prob_poisonous:8.4%}")
    if prob_poisonous >= 0.5:
        print(f"   ⇒ Гриб ЯДОВИТЫЙ ❌ (уверенность {prob_poisonous:.2%})")
    else:
        print(f"   ⇒ Гриб СЪЕДОБНЫЙ ✅ (уверенность {prob_edible:.2%})")

# Реальные примеры
print("1. Нет запаха + коричневый отпечаток спор (типичный шампиньон):")
predict_mushroom({'odor': 5, 'spore-print-color': 2})

print("\n2. Запах foul (вонючий) — почти всегда ядовитый:")
predict_mushroom({'odor': 4})

print("\n3. Запах креозота (creosote) — 100% ядовитый в реальности:")
predict_mushroom({'odor': 2})

print("\n4. Запах миндаля — почти всегда съедобный:")
predict_mushroom({'odor': 0})

print("\n5. Только цвет жабр бурый (buff) — что скажет сеть?")
predict_mushroom({'gill-color': 0})

print("\n6. Ничего не знаем (априорная вероятность):")
predict_mushroom({})

print("\n" + "═" * 80)
````

### 6. Сравнение результатов с baseline-моделью (например, наивный байесовский классификатор из scikit-learn) по метрикам accuracy или log-likelihood.

````python
# === 6. Сравнение с Naïve Bayes из scikit-learn ===

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score, log_loss

X = data.drop('class', axis=1)
y = data['class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Naïve Bayes из sklearn (по умолчанию с лёгким сглаживанием)
nb = CategoricalNB(alpha=1.0)
nb.fit(X_train, y_train)
y_pred_nb = nb.predict(X_test)
y_proba_nb = nb.predict_proba(X_test)

acc_nb = accuracy_score(y_test, y_pred_nb)
ll_nb = log_loss(y_test, y_proba_nb)

print(f"Naïve Bayes (sklearn)  → Accuracy: {acc_nb:.5f}   |   Log-loss: {ll_nb:.5f}")

# Наша байесовская сеть (через inference)
y_pred_bn = []
y_proba_bn = []

for idx, row in X_test.iterrows():
    evidence = row.to_dict()
    q = infer.query(['class'], evidence=evidence)
    prob_p = q.values[1]
    y_pred_bn.append(1 if prob_p > 0.5 else 0)
    y_proba_bn.append([1 - prob_p, prob_p])

acc_bn = accuracy_score(y_test, y_pred_bn)
ll_bn = log_loss(y_test, y_proba_bn)

print(f"Байесовская сеть (pgmpy) → Accuracy: {acc_bn:.5f}   |   Log-loss: {ll_bn:.5f}")

# Итог
print("\n" + "="*60)
print("ИТОГОВОЕ СРАВНЕНИЕ")
print("="*60)
print(f"{'Модель':<25} {'Accuracy':<12} {'Log-loss'}")
print("-"*60)
print(f"{'Naïve Bayes (sklearn)':<25} {acc_nb:.5f}      {ll_nb:.5f}")
print(f"{'Байесовская сеть':<25} {acc_bn:.5f}      {ll_bn:.5f}")
print("="*60)
````

При сравнении построенной байесовской сети с наивным байесовским классификатором из scikit-learn (CategoricalNB) получены следующие результаты:

Кроме того, байесовская сеть даёт значительно более калиброванные вероятности (меньший log-loss), что делает её предсказания не только точнее, но и лучше интерпретируемыми.

В ходе лабораторной работы была построена и исследована байесовская сеть для задачи классификации грибов на съедобные и ядовитые (датасет Mushroom). 

Структура сети была автоматически обучена с использованием алгоритма HillClimbSearch и BIC-оценки, что позволило выявить ключевые зависимости: центральную роль признака odor (запах), а также сильное влияние spore-print-color, gill-color и ring-type на целевую переменную class.

Параметры сети оценены с помощью BayesianEstimator (BDeu-приор), что обеспечило устойчивые вероятностные оценки даже при редких комбинациях признаков. Проведённый вероятностный вывод продемонстрировал высокую интерпретируемость модели: например, наличие запаха foul, creosote или pungent практически однозначно указывает на ядовитость гриба.

Сравнение с наивным байесовским классификатором показало превосходство построенной байесовской сети как по точности классификации (100% против ~96%), так и по качеству вероятностных предсказаний (log-loss 0.00147 против 0.113).
